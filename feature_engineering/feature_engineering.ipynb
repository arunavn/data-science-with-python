### Label Encoding

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
iris_df = pd.read_csv('Iris.csv')
iris_df.shape, iris_df.head(3)
```

```python
#taking info of the database
iris_df.info()
```

```python
# the only column with strig type is species
iris_df.species.unique()
```

```python
## Attempt to encode species
label_encoder = preprocessing.LabelEncoder()
```

```python
label_encoder.fit(iris_df.species.unique())
```

```python
label_encoder.transform(iris_df.species.values)
```

```python
## fit and transform at one go
label_encoder = preprocessing.LabelEncoder()
iris_df['species'] = label_encoder.fit_transform(iris_df.species)
```

### One Hot Encoding

```python
## now let us try to encode species col that has 0, 1 ,2 with one  Hot encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot_encoder = OneHotEncoder(categories='auto')
```

```python
y = iris_df.species.values
y = y.reshape([-1,1])
```

```python
y = onehot_encoder.fit_transform(y)
y.toarray()
```

```python
df[['RelationshipSatisfaction','JobSatisfaction','EnvironmentSatisfaction','JobInvolvement', 'WorkLifeBalance']]
```

```python
df['satisfaction'] = (df['RelationshipSatisfaction']  + df['EnvironmentSatisfaction'] 
                                + df['JobSatisfaction'] + df['JobInvolvement'] + df['WorkLifeBalance'])/5
```

```python
df.head(2)
```

```python
def is_satisfied(df):
    if df['satisfaction'] > 2.35:
        return True
    return False

df['satisfaction_flag'] = df.apply(lambda df: is_satisfied(df), axis = 1)
```
